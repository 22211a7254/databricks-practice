```python
# Create sample data
data = [
    ("Laxmi", 21, "Hyderabad"),
    ("Ravi", 25, "Bangalore"),
    ("Anu", 23, "Chennai"),
    ("Kiran", 28, "Mumbai")
]

columns = ["Name", "Age", "City"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Show data
df.show()

# Filter data
filtered_df = df.filter(df.Age > 23)

print("People with Age > 23:")
filtered_df.show()

# Add new column
from pyspark.sql.functions import col

df2 = df.withColumn("Age_in_5_years", col("Age") + 5)

print("Updated Data:")
df2.show()
```
